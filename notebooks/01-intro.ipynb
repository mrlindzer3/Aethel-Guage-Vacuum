# Aethel Lab — Intro
This notebook runs a tiny deterministic experiment using core.solid_state_lab and plots a tiny result.

```python
from core.solid_state_lab.lattice import Lattice
from core.solid_state_lab.optomech import OptomechanicalSystem
from core.solid_state_lab.experiments import Experiment

lat = Lattice(lattice_constant=1.0, dims=1, size=8)
sys = OptomechanicalSystem(lat)
exp = Experiment(name='notebook-demo', system=sys)
metrics = exp.run(cycles=5, seed=0)
print(metrics)
```
